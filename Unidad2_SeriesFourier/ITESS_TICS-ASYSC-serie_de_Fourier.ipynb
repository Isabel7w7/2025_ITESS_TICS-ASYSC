```python
import os
from thinkdsp import CosSignal, SinSignal
import numpy as np
#a1
k = 1 
PI = np.pi

cos_signal = CosSignal(freq=0.25, amp=2*k/PI, offset=0)
wave = cos_signal.make_wave(duration = 10, framerate = 100)
wave.ys += 0.5
  
```

```python
from thinkdsp import decorate
import matplotlib.pyplot as plt

wave.plot()
decorate(xlabel='Time(s)')
plt.grid (True, which ='both')
```
